```python
import numpy as np
import pandas as pd
import lightgbm as lgb
import xgboost as xgb
from sklearn.linear_model import BayesianRidge
from sklearn.model_selection import KFold, RepeatedKFold
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from scipy import sparse
import warnings
import time
import sys
import os
import re
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
from sklearn.metrics import mean_squared_error
from sklearn.metrics import log_loss
from dateutil.parser import parse
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns',None)
pd.set_option('max_colwidth',100)

train = pd.read_csv('datalab/7955/jinnan_round1_train_20181227.csv', encoding = 'gb18030')
test  = pd.read_csv('datalab/7955/jinnan_round1_testA_20181227.csv', encoding = 'gb18030')
# 删除异常值
train = train[train['收率'] > 0.87]
# 计算时间差
def TimeDiff(szTime1, szTime2):
    d = 0
    if "24:00:00" == szTime1:
        szTime1 = "23:59:59"
        d = -1
    if "24:00:00" == szTime2:
        szTime2 = "23:59:59"
        d = d + 1
    a = parse(szTime1)
    b = parse(szTime2)
    timeDiff = b-a
    if timeDiff.days == -1:
        return TimeDiff(szTime1, "23:59:59") + TimeDiff("00:00:00", szTime2) + 1 + d
    return timeDiff.seconds + d
```

```python
# 合并数据集
target = train['收率']
del train['收率']
data = pd.concat([train,test],axis=0,ignore_index=True)
data["id"] = data["样本id"]
data.shape
```

```python
#分离时间段
IntervalT_index = ["A20","A28","B4","B9","B10","B11"]
for col in IntervalT_index:
    col_per = col+"_per"
    col_last = col+"_last"
    data[col_per] = data.A1
    data[col_last] = data.A1
    for i in data.index:
        intervalTmp = data.loc[[i], [col]]
        #print(intervalTmp)
        if pd.isnull(intervalTmp.values[0][0]):
            data.loc[[i], [col_per]] = data.loc[[1],["A2"]]
            data.loc[[i], [col_last]] = data.loc[[1],["A2"]]
        else:
            splitList = intervalTmp.values[0][0].split("-")
            col_per_value = splitList[0]
            col_last_value = splitList[1]
            data.loc[[i], [col_per]] = col_per_value+ ":00"
            data.loc[[i], [col_last]] = col_last_value+ ":00"
print("结束")
```

# 计算时间间隔

```python
T_col = ["A5","A7","A9","A11","A14","A16",
         "A20_per","A20_last","A24","A26","A28_per",
         "A28_last","B4_per","B4_last","B5","B7",
         "B9_per","B9_last","B10_per","B10_last","B11_per","B11_last"]
```

```python
data["oTmp"] = data.A5
try:
    for i1 in range(len(T_col)):
        col = T_col[i1]
        print(col)
        col_o = col + "_o"
        data[col_o] = data.A1
        for i in data.index:
            #print(i)
            T_Tmp = data.loc[[i], [col]].values[0][0]
            O_Tmp = data.loc[[i], ["oTmp"]].values[0][0]
            if pd.isnull(T_Tmp):
                data.loc[[i], [col_o]] = 0
            else:
                data.loc[[i], [col_o]] = TimeDiff(T_Tmp, O_Tmp)
                data.loc[[i], ["oTmp"]] = T_Tmp
except:
    print(i1,i,T_Tmp,O_Tmp,data.loc[i]["样本id"])
print("结束")
```

```python
T_col_drop = ["A7","A9","A11","A14","A16",
         "A20","A20_per","A20_last","A24","A26","A28","A28_per",
         "A28_last","B4","B4_per","B4_last","B5","B7",
         "B9","B9_per","B9_last","B10","B10_per","B10_last","B11","B11_per","B11_last",
         "oTmp","样本id"]
data.drop(T_col_drop, axis=1, inplace=True)
```
